```python
import pandas as pd

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score

from wordcloud import WordCloud

import matplotlib.pyplot as plt
```

```python
nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

stemmer = PorterStemmer()
```

```python
data.shape
```

```python
def combine_2rd_columns(col_1, col_2, col_3):
    result = col_1
    if not pd.isna(col_2):
        result += " " + str(col_2)
    return result


data = pd.read_csv("pilot_textming.csv")
data["text"] = data.apply(lambda x: combine_2rd_columns(x['Title'], x['Keywords'], x['Abstract']), axis=1)

df = data.loc[:, ['text']]
df.head()
```

#### Data Preprocess

```python
def preprocess_text(text):
    # 소문자 변환 및 토큰화 (단어 단위로 분할)
    tokens = text.lower().split()
    
    # Stopword 제거 및 Lemmatization 수행
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words]
    
    # Stemming 수행 (Porter Stemmer 사용)
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(token) for token in tokens]
    
    # 숫자 제거 및 재결합하여 문자열 반환
    processed_text = ' '.join([token for token in tokens if not token.isdigit()])
    
    return processed_text

df['processed_text'] = df['text'].apply(preprocess_text)
df = df.loc[:, ['processed_text']]
df.head()
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['processed_text'])
tfidf_matrix
```

#### Selected only the word-features - based on TF-IDF(Python scikit-Learn Library)

```python
# 상위 Docuement 당 TF-IDF 기반 상위 4개의 단어 추출 함수 정의

top_keywords_per_document = []
feature_names = vectorizer.get_feature_names_out()

for i in range(len(df)):
    tfidf_scores = zip(feature_names, tfidf_matrix[i].toarray()[0])
    sorted_tfidf_scores = sorted(tfidf_scores,key=lambda x: x[1], reverse=True)
    top_keywords_per_document.append([keyword for keyword,_ in sorted_tfidf_scores[:4]])

# 전체 데이터셋에서 적어도 2번 이상 나온 키워드 추출하기 
all_keywords=[keyword for document in top_keywords_per_document for keyword in document]
selected_keywords=list(set(keyword for keyword in all_keywords if all_keywords.count(keyword) >=2))

# 선택된 키워드를 기반으로 Document-Term Matrix 생성
vectorizer=CountVectorizer(vocabulary=selected_keywords)
document_term_matrix=vectorizer.fit_transform(df['processed_text'])

# WordCloud 시각화 함수 정의 및 실행   
def visualize_wordcloud(keywords):
     wordcloud=WordCloud(background_color='white', width=800,height=400).generate(' '.join(keywords))
     plt.imshow(wordcloud.recolor(colormap='viridis',random_state=42),interpolation="bilinear")
     plt.axis("off")
     plt.show()

visualize_wordcloud(selected_keywords)
document_term_matrix
```

```python
# 재정규화된 TF-IDF 값으로 단어 분포 확인

renormalized_tfidf_matrix = tfidf_matrix / tfidf_matrix.sum(axis=1)
word_distribution_per_document = renormalized_tfidf_matrix.toarray()

print("\nWord Distribution per Document:")
for i in range(len(df)):
    print(f"Document {i+1}:")
    word_distribution_pairs = zip(feature_names, word_distribution_per_document[i])
    sorted_word_distribution_pairs = sorted(word_distribution_pairs,key=lambda x: x[1], reverse=True)
    
    for word,distribution in sorted_word_distribution_pairs:
        print(f"{word}: {distribution}")
```
